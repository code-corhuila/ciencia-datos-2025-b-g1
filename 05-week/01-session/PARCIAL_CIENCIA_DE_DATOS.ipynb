# An√°lisis de la Relaci√≥n entre Actividades y Parciales

## 1. Objetivo del an√°lisis
El prop√≥sito de este trabajo fue **analizar la relaci√≥n entre los promedios de las actividades y los promedios de los parciales de los estudiantes**, con el fin de determinar si aquellos que obtienen buenos resultados en las actividades tambi√©n tienden a lograr un buen desempe√±o en los parciales.  
Este an√°lisis se realiz√≥ a nivel **grupal**, es decir, considerando el comportamiento general del conjunto de estudiantes y no solo de manera individual.

---

## 2. Datos y variables utilizadas
El dataset contiene informaci√≥n de las notas de diferentes componentes acad√©micos (Talleres, Foros, Quices, Parciales, etc.).  
A partir de estos datos, se construyeron las siguientes variables principales:

- **Actividades_Final**: corresponde al promedio de todas las actividades (Talleres, Foros y Quices) de los tres cortes.  
- **Parcial_Final**: corresponde al promedio de los tres parciales (uno por cada corte).

Estas dos variables fueron las **principales de an√°lisis** para establecer la relaci√≥n entre desempe√±o en actividades y desempe√±o en parciales.

---

## 3. Metodolog√≠a
1. Se importaron los datos desde un archivo CSV con ayuda de la librer√≠a `pandas`.
2. Se calcularon las variables **Actividades_Final** y **Parcial_Final** a partir de las notas de cada estudiante.
3. Se aplicaron dos enfoques de visualizaci√≥n:
   - Un **gr√°fico de dispersi√≥n**, para observar el comportamiento de cada estudiante y la relaci√≥n entre ambas variables.
   - Un **gr√°fico de barras comparativo**, para analizar los promedios globales del grupo en Actividades y Parciales.
4. Se interpretaron los resultados considerando el rendimiento colectivo.

---

## 4. Visualizaciones

### 4.1. Gr√°fico de Dispersi√≥n
El gr√°fico de dispersi√≥n representa a cada estudiante como un punto en el plano:  
- El eje X corresponde al **promedio de actividades**.  
- El eje Y corresponde al **promedio de parciales**.  
- Cada punto tiene un color diferente para mejorar la visualizaci√≥n y distinguir los estudiantes.

Este gr√°fico se utiliz√≥ porque permite observar si existe una **tendencia o patr√≥n de relaci√≥n** entre ambas variables. En este caso, muestra la distribuci√≥n de estudiantes y si los que tienen altas calificaciones en actividades tambi√©n logran altas calificaciones en parciales.

---

### 4.2. Gr√°fico de Barras Comparativo
El gr√°fico de barras presenta el **promedio general del grupo** en las dos variables:  
- Una barra para **Actividades_Final**.  
- Una barra para **Parciales_Final**.  

Este gr√°fico permite comparar de manera sencilla el **nivel promedio alcanzado por el grupo** en cada componente. La elecci√≥n de este tipo de gr√°fico se debe a que facilita la comparaci√≥n directa de promedios.

---


## 5. Conclusiones

1. El an√°lisis permiti√≥ identificar que, en promedio, los estudiantes obtienen mejores resultados en **actividades** (talleres, foros y quices) que en los **parciales**, lo cual sugiere que estos √∫ltimos representan un mayor nivel de exigencia.  

2. El **gr√°fico de dispersi√≥n** muestra que no siempre existe una relaci√≥n directa: algunos estudiantes con buen desempe√±o en actividades no logran los mismos resultados en los parciales. Esto indica que las actividades, tal como est√°n dise√±adas, **no siempre garantizan un refuerzo suficiente** para enfrentar las evaluaciones finales.  

3. El **gr√°fico de barras comparativo** refuerza esta conclusi√≥n, ya que evidencia que el promedio grupal en actividades supera al de parciales, marcando una diferencia clara entre ambos componentes.  

4. A nivel institucional, este tipo de an√°lisis es valioso porque permite evaluar si las actividades realmente contribuyen al aprendizaje o si deben replantearse para estar m√°s alineadas con los objetivos de los parciales.  

5. No se trata √∫nicamente de aumentar la cantidad de talleres, sino de revisar su **calidad y pertinencia**. Actividades bien dise√±adas y relacionadas con los contenidos evaluados pueden convertirse en un apoyo real para mejorar el rendimiento acad√©mico en los parciales.  


```python
import pandas as pd # ESTA ES UNA LIBRERIA PARA IMPORTAR ARCHIVOS DE EXCEL
import matplotlib.pyplot as plt # ESTA ES LA LIBRERIA PARA GRAFICAR
import numpy as np #ESTA ES LA LIBRERIA PARA HACER CACLULOS

# 1. Montar Drive
from google.colab import drive #ESTO ES PARA CONECTAR CON MI DRIVE
drive.mount('/content/drive')

# 2. Leer archivo
ruta = "/content/drive/MyDrive/Ciencia-datos-2025-b-g1/dataset_notas.csv"
df = pd.read_csv(ruta) # ESTA ES LA RUTA QUE TENGO CON MI DRIVE

# 3. Crear variables finales
#df es la variable para llamar los datos del excel
# aqui utilizamos la formula de sumar todas las actividades y dividirlas en la cantidad
df["Actividades_Final"] = (
    df["Taller_1"] + df["Foro_1"] + df["Quiz_1"] +
    df["Taller_2"] + df["Foro_2"] + df["Quiz_2"] +
    df["Taller_3"] + df["Foro_3"] + df["Quiz_3"]
) / 9   # promedio de todas las actividades por estudiante

# esta es la variable de parcial final, la cual suma todos los datos de cada parcial
#luego se suman los parciales de cada corte y se dividen en 3, por que son 3 cortes
df["Parcial_Final"] = (df["Parcial_1"] + df["Parcial_2"] + df["Parcial_3"]) / 3


# 4. Gr√°fico de dispersi√≥n con colores por estudiante
# esta funcion la busque en chat gpt, sirve para que a cada estudiante
#en la grafica de dispersion le de un color
# la funcion len () sirve para saber cuantas filas hay
# la funcion np.random.rand () sirve para que me de un n√πmero aleatroio
#entre 0 y 1 y da una escala de colores
colores = np.random.rand(len(df))  # colores aleatorios para cada estudiante

# esto es para crear la hoja de dibujo en este caso es de 10*6 pulgadas
plt.figure(figsize=(10,6))
# plt.scatter () esto es para que se cree un punto por cada estudiante
#en phyton la primera variable siempre es la X, osea actividades
#la segunda es Y, osea parcial
scatter = plt.scatter(df["Actividades_Final"], df["Parcial_Final"],
                      c=colores, cmap="rainbow", s=80, edgecolor="black", alpha=0.7)

#las """ se usan para comentarios de varias lineas
"""
Para la configuracion de colores me apoye de IA, pues no sabia del tema
c=colores ‚Üí el color de cada punto depende del n√∫mero aleatorio que sali√≥ antes.
cmap="rainbow" ‚Üí usa la paleta de colores tipo arco√≠ris.
s=80 ‚Üí tama√±o de los puntos.
edgecolor="black" ‚Üí cada punto tiene borde negro.
alpha=0.7 ‚Üí los puntos son un poquito transparentes.
"""

# esto es para poner los nombres en los ejes
plt.xlabel("Promedio Actividades Final") # Nombre del eje X
plt.ylabel("Promedio Parciales Final") # Nombre dele eje Y
#titulo General del grafico
plt.title("Relaci√≥n entre Actividades y Parciales (Promedios Finales)")
plt.xlim(0,5) #Limite de la grafica en X
plt.ylim(0,5) # limite de la grafica en eje Y


plt.show()

# 5. Gr√°fico de barras comparando promedios globales
# se toma toda la columna de estudiantes de la variable actividades final
# y por medio de la funci√≤n mean se saca el promedio del curso
prom_actividades = df["Actividades_Final"].mean()
#aqui tambein se saca el promedio pero de la columna parcial Final
prom_parciales = df["Parcial_Final"].mean()


# Nombres que apareceran en el eje X, habran dos barras
etiquetas = ["Actividades Final", "Parciales Final"]

# esta es la altura de cada barra, que seran los promedios
valores = [prom_actividades, prom_parciales]

# esto lo saque de chat gpt y son los colores de cada barra, para un azul clarito y una salm√≥n
colores_barras = ["skyblue", "salmon"]

#la funcion plt.bar es para un grafico de barras

plt.bar(etiquetas, valores, color=colores_barras)

plt.ylim(0,5) #En el eje Y va de 0 a 5
plt.ylabel("Promedio") # Nopmbre en el eje Y
plt.title("Comparaci√≥n Promedios Finales Actividades vs Parciales") #Nombre del grafico
plt.show()  # para mostrar el grafico
```
